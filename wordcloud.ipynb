```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Baca file CSV
file_path = 'quran_id_202408240608.csv'  
df = pd.read_csv(file_path)

# Ambil teks dari kolom 'indotext'
text = ' '.join(df['indoText'].dropna().tolist())

# Buat word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='viridis').generate(text)

# Tampilkan word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
import nltk

# Pastikan untuk mendownload stopwords dan punkt jika belum
nltk.download('stopwords')
nltk.download('punkt')

# Baca file CSV
file_path = 'quran_id_202408240608.csv'   # Ganti dengan path file kamu
df = pd.read_csv(file_path)

# Ambil teks dari kolom 'indoText'
text = ' '.join(df['indoText'].dropna().tolist())

# Preprocessing
# 1. Lowercasing
# text = text.lower()

# 2. Remove punctuation
text = text.translate(str.maketrans('', '', string.punctuation))

# 3. Tokenization
words = word_tokenize(text)

# 4. Remove stop words (kata umum bahasa Indonesia)
stop_words = set(stopwords.words('indonesian'))
words = [word for word in words if word not in stop_words]

# 5. Bergabung kembali menjadi satu string
processed_text = ' '.join(words)

# Buat word cloud dari teks yang sudah dipreprocessing
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(processed_text)

# Tampilkan word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
