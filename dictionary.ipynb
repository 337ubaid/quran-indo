```python
import csv

dict = {}
with open('quran_id_full.csv', mode='r', encoding='utf-8') as file:
    data_quran = csv.DictReader(file)

    for ayat in data_quran:
        key = (int(ayat['suraId']), int(ayat['verseID']))
        value = ayat['indoText']

        dict[key] = value
```

```python
print(dict[(114,1)])
```

```python
for key, value in dict.items():
    print(f"Key: {key}, Value: {value}")
```
