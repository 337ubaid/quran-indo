```python
model_name = '../sbert/search/search-model'
index_name = "../sbert/quran_embed_finetune.index"
```

```python
from sentence_transformers import SentenceTransformer
import faiss
```

```python
model = SentenceTransformer(model_name)
index= faiss.read_index(index_name)
```

```python
query_vector = model.encode(["apa itu ikhlas"])
query_vector
```

```python
top_5 = index.search(query_vector, 5)
top_5
```
